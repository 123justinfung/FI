```python
#USD/Base Ccy
import pandas as pd
currnecies_list = ['EUR=X', 'GBP=X', 'JPY=X', 'CNY=X', 'EUR=X', 'AUD=X', 'NZD=X', 'CAD=X', 'CHF=X', 'NOK=X', "SEK=X"]
currencies = yf.Tickers(currnecies_list)
df  = pd.DataFrame(currencies.history(start='2006-06-01', end='2023-05-01')['Close'])
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
df_rv = df
for col in df_rv.columns[1:]:
    df_rv[col] = df_rv[col]/df_rv[col].iloc[0]
df_rv.reset_index()
print(df_rv)
```
